```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Step 2: Use requests and BeautifulSoup to make a list of all the CORGIS datasets.
# Use requests with the following site: "https://corgis-edu.github.io/corgis/csv/"
# The relevant HTML tag to search for is "h3"
html_doc = requests.get("https://corgis-edu.github.io/corgis/csv/").text
soup = BeautifulSoup(html_doc, 'html.parser')
h3 = soup.find_all('h3')
# The list of all CORGIS datasets
datasets = []
for heading in h3:
    datasets.append(heading.string)
```

```python
# Step 3: Write a function that takes an element from the list of CORGIS datasets, 
# searches the respective CORGIS page for the CSV download link, and returns a Pandas dataframe.
def pdcorgis(name):
    dataset = name.lower()
    html_doc = requests.get("https://corgis-edu.github.io/corgis/csv/" + dataset + "/").text
    soup = BeautifulSoup(html_doc, 'html.parser')
    a = soup.find_all('a')
    for link in a:
        if link.has_attr('download'):
            url = "https://corgis-edu.github.io/corgis/csv/" + dataset + "/" + link['href']
            return pd.read_csv(url)
```

```python
# Step 4: Using dataframes returned by your new function, make a line plot, a bar plot, and a histogram plot
video_games = pdcorgis("Video_Games")
opioids = pdcorgis("Opioids")
cancer = pdcorgis("Cancer")

#video_games.head()
#opioids.head()
#cancer.head()

fig, axs = plt.subplots(3, 1, figsize=(10,35))

ax1, ax2, ax3 = axs

# Line chart
opioids.plot(kind="line", x="Year", y="Number.Opioid.Any", ax = ax1)

# Bar chart
cancer.plot(kind="barh", x="State", y="Total.Number", color="red", ax = ax2)

# Histogram chart
video_games["Metrics.Review Score"].plot.hist(color="green", ax = ax3, xlabel="Review Score")
```
